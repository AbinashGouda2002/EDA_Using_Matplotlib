```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
d={"Length":np.random.randn(500)}
df=pd.DataFrame(d)
df
```

```python
df.shape
```

```python
df.size
```

```python
#Histogram
df["Length"].hist()
```

```python
plt.hist(df["Length"]) # For bins will be write as plt.hist(df["Length"],bins=10)
plt.xlabel("Length")  # named as x---> axis
plt.ylabel("count/frequency") # named as y---> axis
plt.title("Histogram of length") # Title named as Histogram of length
plt.show()
```

```python
plt.hist(df["Length"],bins=15)
plt.show()
```

```python
plt.hist(df["Length"],orientation='horizontal')
plt.xlabel("count")
plt.ylabel("length")
plt.title("Histogram of length")
plt.show()
```

```python
plt.hist(df["Length"],orientation="horizontal")
plt.xlabel("Count")
plt.ylabel("Length")
plt.title("Histogram")
plt.show()
```

```python
# Box Plot
plt.boxplot(df["Length"])
plt.show()
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
plt.plot(days,max_t,label="Maximum")
plt.plot(days,min_t,label="Minimum")
plt.plot(days,avg_t,label="Average")
plt.xlabel("Days")
plt.ylabel("Weather")
plt.title("Weekly Temperature")
plt.legend()
plt.show
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
plt.subplot(2,2,1)
plt.plot(x,y, color="green",linestyle="dashed",marker="*")
plt.title("line plot")

plt.subplot(2,2,2)
plt.scatter(x,y,color="green",marker="^")
plt.title("Scatter plot")

plt.subplot(2,2,3)
plt.scatter(x,y)
plt.xlabel("days")
plt.ylabel("temperature")

plt.subplot(2,2,4)
plt.plot(x,y,color="red",marker="o",linewidth="3")
plt.title("2nd lineplot")
plt.savefig("lineplot.pdf")
plt.show()


```

```python
list=[20,30,15,10,20,5]
lbs=["FaceCream","FaceWash","ToothPaste","Bathing Soap","sampoo","Moisturizer"]
```

```python
la='python','java','c++','R'
size=[50,35,10,5]
colour=["gold","yellowgreen","lightcoral","lightskyblue"]
explode=[0.1,0,0,0]
plt.pie(size,labels=la,autopct='%1.1f%%',explode=explode,radius=1.5,colors=colour)
plt.savefig("barplot.pdf")
plt.show()
```

```python
company=["google","microsoft","amazone","facebook"]
revenue=[90,89,136,27]
plt.bar(company,revenue)
plt.ylabel("revenue($b)")
plt.title("US Techonology Stocks")
plt.show()
```

```python
plt.barh(company,revenue) #plt.barh--->"h is most important for horizontal"
plt.ylabel("revenue(Billion)")
plt.title("US Technology Stocks")
plt.show()
```

```python
#Stacked bar plot
company=["google","microsoft","amazone","facebook"]
Revenue=[123,70,90,88]
profit=[20,25,35,50]
plt.bar(company,Revenue,label="Revenue")
plt.bar(company,profit,label='Profit')
plt.legend()
plt.title("US Technology Stocks")
plt.show()
```

```python
#Revenue wise Google is so high and profit wise facebook is so high
```

```python
company=["google","microsoft","amazone","facebook"]
Revenue=[123,70,90,88]
profit=[20,25,35,50]
xpos=np.arange(len(company))
plt.bar(xpos-0.2,Revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")
plt.xticks(xpos,company)
plt.title("US Technology Stocks")
plt.legend()
plt.figure(1,figsize=(10,10))
plt.show()
```

```python
ecommerce=["myntra","snapdeal","alibaba","amazone","flipkart"]
q1_profit=[35,45,100,70,40]
q2_profit=[38,40,105,65,45]
q3_profit=[38,42,120,72,50]
q4_profit=[25,34,115,60,48]

plt.figure(1,figsize=(10,10))

plt.subplot(2,2,1)
plt.bar(ecommerce,q1_profit)
plt.title("q1 profit")

plt.subplot(2,2,2)
plt.bar(ecommerce,q2_profit)
plt.title("q2 profit")

plt.subplot(2,2,3)
plt.bar(ecommerce,q3_profit)
plt.title("q3 profit")

plt.subplot(2,2,4)
plt.bar(ecommerce,q4_profit)
plt.title("q4 profit")

plt.suptitle("profit of companies at a quater") # always we use plt.suptitle("name")
plt.show()
```
